```python
# import libraries 
import requests
import urllib.request
import time 
from bs4 import BeautifulSoup 

#check the response status of the url 
url = 'http://web.mta.info/developers/turnstile.html'
response = requests.get(url)
response

#parse the html code from the website 
soup = BeautifulSoup (response.text, "html.parser")

#find all the relevant urls which contains specific pattern and store them in a list 
import re
urls =[]
for link in soup.find_all('a', href=re.compile ("data/")):
    urls.append('http://web.mta.info/developers/'+link['href'])
    
#create a range of numbers within that length of urls to use later in the for loop iteration
ra = range(len(urls))

#using for loop to download each file from the website 
for i,j in zip(urls,ra):
    #print the link 
    print(i)
    #find the relevant portion of the url to use it as a file name to save
    file_name_start_pos = i.rfind("/turnstile_") + 1
    file_name = i[file_name_start_pos:-4]
    #get the url and download the file
    urllib.request.urlretrieve(i,"C:/Kaggle/Web/"+str(j)+ file_name +".csv")
    #print the index of the url
    print(j)
    # one second pause in betweent the subsequent request 
    time.sleep(1)
```
